```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = x**2

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$y = x^2$', color='b')

# Customize the plot
plt.title("Line Plot of $y = x^2$")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```

```python
# Data for the bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Create the bar plot
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=colors)

# Customize the plot
plt.title("Bar Plot of Categories")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show the plot
plt.show()
```

```python
# Data for the scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='blue', marker='o')

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

    # Customize the plot
plt.title("Scatter Plot with Annotations")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

```python
# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140)

# Customize the plot
plt.title("Programming Language Popularity")

# Show the plot
plt.show()
```

```python
# Create a figure with a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Data for the plots
x = np.linspace(0, 10, 100)
y = x**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Line plot
axes[0, 0].plot(x, y, color='blue', label=r'$y = x^2$')
axes[0, 0].set_title("Line Plot")
axes[0, 0].set_xlabel("X")
axes[0, 0].set_ylabel("Y")
axes[0, 0].grid(True, linestyle='--', alpha=0.6)
axes[0, 0].legend()

# Bar plot
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title("Bar Plot")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# Scatter plot
axes[1, 0].scatter(scatter_x, scatter_y, color='blue', marker='o')
for i in range(len(scatter_x)):
    axes[1, 0].annotate(f"({scatter_x[i]}, {scatter_y[i]})", (scatter_x[i], scatter_y[i]), 
                             textcoords="offset points", xytext=(5,5), ha='center')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X Values")
axes[1, 0].set_ylabel("Y Values")
axes[1, 0].grid(True, linestyle='--', alpha=0.6)
 # Pie chart
axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140)
axes[1, 1].set_title("Pie Chart")

                             # Adjust layout for better spacing
plt.tight_layout()

                             # Show the plot
plt.show()
    
```

```python
# Data for the histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create the histogram
plt.figure(figsize=(6, 4))
plt.hist(data, bins=5, color='teal', edgecolor='black', alpha=0.7)

# Customize the plot
plt.title("Histogram of Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```
